### Supervised Leaning Approach 

```python
y = df['Response'] ## dependent variable
x_new =df.drop(['Response', 'Education'] ,axis =1)  ## independent variable 
```

```python
from sklearn.model_selection import train_test_split
x_train , x_test , y_train , y_test = train_test_split(x_new, y, test_size =0.2 , random_state =51)

print('Shape of x_train = ' , x_train.shape)
print('Shape of y_train = ' , y_train.shape)
print('Shape of x_test = ' , x_test.shape)
print('Shape of y_test = ' , y_test.shape)
```

### Decision Tree 

```python
from sklearn.tree import DecisionTreeClassifier
classifier = DecisionTreeClassifier(criterion ='gini')
classifier.fit(x_train, y_train)
```

```python
classifier.score(x_test, y_test)
```

#### KNN

```python
from sklearn.neighbors import KNeighborsClassifier
classifier = KNeighborsClassifier(n_neighbors=5)
classifier.fit(x_train ,y_train)
classifier.score(x_test,y_test)
```

### Random Forest
